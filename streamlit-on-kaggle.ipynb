```python
%%writefile app.py
import streamlit as st

st.title('Hello Streamlit')
st.write('Welcome to your Streamlit app running in a Kaggle notebook!')
```

```python
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
ngrok_secret_value = user_secrets.get_secret("ngrok")
```

```python
from pyngrok import ngrok
import subprocess
ngrok.set_auth_token(ngrok_secret_value)
# Define the Streamlit app file
app_file = "app.py"


process = subprocess.Popen(["streamlit", "run", app_file], stdout=subprocess.PIPE)

# Create a tunnel to the Streamlit app
public_url = ngrok.connect(8501, "http")
print(f"Streamlit app running at {public_url}")
```

```python
# To stop the Streamlit app and ngrok tunnel:
# Terminate the Streamlit app process
process.terminate()

# Disconnect the ngrok tunnel
ngrok.disconnect(public_url)

print("Streamlit app and ngrok tunnel stopped.")
```
